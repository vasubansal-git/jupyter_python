# Importing in Python is the process of bringing code from one module (file) into another so you can use its functions, classes, or variables.A module is simply a .py file that contains Python code. When you import a module:

- Python searches for the module.

- It loads and executes the module code (only once).

- It stores it in memory.

## You can then use its contents in your program.

```python
import math
math.floor(2.89)
```

```python
result = math.sqrt(9)
```

```python
print(result)
```

```python
from math import sqrt, pi
result = sqrt(10)
print(result)
```

```python
result1 = sqrt(12) * pi
print(result1)
```

```python
# from math import * (this is generally not recommended)

from math import sqrt as s, pi as p

result2 = s(9) * p
print(result2)
```
