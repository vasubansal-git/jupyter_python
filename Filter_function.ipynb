# Filter Function in Python

```python
def even(num):
    if num % 2 == 0:
        return True
```

```python
even(4)
```

```python
lst = [1, 2, 3, 4, 5, 6, 7, 8, 9, 0]
```

```python
filter(even, lst)
```

```python
list(filter(even, lst))
```

```python
list(filter(lambda num: num % 2 == 0, lst))
```

```python
list(map(lambda num: num%2 == 0, lst))
```

```python
lst1 = [10, 20, 30, 11, 22, 33, 45, 67, 87, 90, 100]
list(filter(lambda x: x > 20, lst1))
```

```python
list(filter(lambda x: x >= 20, lst1))
```
