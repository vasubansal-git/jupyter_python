# Exception Handling
## Exception handling is the process of responding to unwanted o unexpected events when a computer program runs. Exception handling deals with these events to avoid the program or system crashing and without this process, exceptions would disrupt the normal operation of a program

```python
a = int(input("Enter the number: "))
print(f"Multipliation table of {a} is: ")
for i in range(1, 11):
    print(f"{a} X {i} = {a * i}")

print("Some lines of code")
print("End of program")
```

```python
# try & Except block

a = input("Enter the number: ")
print(f"Multipliation table of {a} is: ")
try:
    for i in range(1, 11):
        print(f"{int(a)} X {i} = {int(a) * i}")
except Exception as e:
    # print(e)
    print("Sorry some error occured")
    
print("Some lines of code")
print("End of program")
```

```python
try:
    num = int(input("Enter an integer: "))
    a = [6, 3]
    print(a[num])
except ValueError:
    print("Number entered is not an integer")
except IndexError:
    print("Index Error")
```
