# Sets

# Dictionaries
## A dictionary is a collection which is unordered, changeable and indexed. IN python dictionaries are written with curly brackets, and they have keys and values

```python
dic = {}
print(type(dic))
```

```python
dic = {1, 2, 3, 4, 5, 6}
print(type(dic))
```

```python
dic1 = {"Name": "Vasu", "Roll No.": "11242502", "Course":"Btech(CSE)"}
print(type(dic1))
```

```python
print(dic1)
```

```python
#Access the values based on Keys
dic1["Name"]
```

```python
# We can even Loop throught the dictionaries keys
for details in dic1:
    print(details)
```

```python
# We can even Loop throught the dictionaries Values
for x in dic1.values():
    print(x)
```

```python
# You can also check both keys and values
for x in dic1.items():
    print(x)
```

```python
# Adding items in dictionaries

dic1["Specialization"] = "AI / ML"
print(dic1)
```

```python
for x in dic1.items():
    print(x)
```

## Nested Dictionary

```python
car1_model = {"Mercedes": 1960}
car2_model = {"Audi": 1970}
car3_model = {"Ambassadar": 1980}

car_type = {"car1": car1_model, "car2": car2_model, "car3": car3_model}
```

```python
print(car_type)
```

```python
# Accessing the items in the dictionaries

print(car_type["car1"])
```

```python
print(car_type["car1"] ["Mercedes"])
```

```python
# We can change whole list of the tuple but cannot be change a single item of tuple
```

```python
# To change the single element in tuple with the help of list
number_tuple = (10, 20, 30, 40, 50)

number_list = list(number_tuple)
number_list[0] = 100
number_tuple = tuple(number_list)
```

```python
print(number_tuple)
```

```python
# To check the element is present in the tuple?
number_tuple.index(100)
```
