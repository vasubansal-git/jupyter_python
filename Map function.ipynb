# Map function
- map() function in Python applies a given function to each element of an iterable (list, tuple, set, etc.) and returns a map object (iterator). It is a higher-order function used for uniform element-wise transformations, enabling concise and efficient code.

```python
def even_odd(n):
    if n % 2 == 0:
        return "The number {} is Even".format(n)
    else:
        return "The number {} is Odd".format(n)
```

```python
even_odd(24)
```

```python
lst = [1, 2, 3, 4, 5, 6, 7, 8, 9, 12, 14, 15, 89, 88, 90]
list(map(even_odd,lst)) # map(func_name, arguments)
```

```python
# Convert string to int using map function

numbers = ["1", "2", "3", "4", "5"]

result = list(map(int, numbers))
print(result)
```

```python
# Find cube of numbers using map function

numbers2 = [1, 2, 3, 4, 5, 6]

cube = list(map(lambda x: x ** 3, numbers2))
print(cube)

# Find square of numbers
square = list(map(lambda x: x ** 2, numbers2))
print(square)
```

```python
# Convert lower case string to upper case using map fuction

strings = ["vasu", "ujjwal", "aman", "karan"]

result = list(map(str.upper, strings))
print(result)
```

```python
# Add two lists using map function

lst1 = [1, 2, 3, 4]
lst2 = [5, 6, 7, 8]

result = list(map(lambda x, y: x + y, lst1, lst2))
print(result)
```
