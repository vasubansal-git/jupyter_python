```python
x = int(input("Enter the value of x: "))
match x:
    case 0:
        print("X is zero")
    case 1:
        print("X is One")
    case 2:
        print("X is Two")
    case 3:
        print("X is Three")
    case 4:
        print("X is Four")
    case _:
        print("Invalid!")
```

```python
def check_vowel(x):
    match x:
        case 'a' | 'e' | 'i' | 'o' | 'u':
            print(x, "is a vowel")
        case _:
            print(x, "is not a vowel")
```

```python
check_vowel('a')
```

```python
check_vowel('o')
```

```python
check_vowel('p')
```

```python
def checkNumber(num):
    match num:
        case num if num > 0:
            print(num, "is a positive number")
        case num if num < 0:
            print(num, "is a negative number")
        case _:
            print(num, "is zero")
```

```python
checkNumber(0)
```

```python
checkNumber(67)
```

```python
def calculator(a, b, op):
    match op:
        case '+':
            print("Addition is performed: ", a + b)
        case '-':
            print("Subtraction is performed: ", a - b)
        case '*':
            print("Multiplication is performed: ", a * b)
        case '/':
            print("Division is performed: ", a / b)
        case _:
            print("Invalid operation, please try again!")
```

```python
calculator(5, 6, '+')
```

```python
calculator(6, 5, '-')
```

```python
calculator(5, 6, '%')
```
