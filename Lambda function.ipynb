# LAMBDA Function
- A lambda function is a small anonymous function.
- A lambda function can take any number of arguments, but can only have one expression.

```python
def addition(a, b):
    return a + b
```

```python
addition(4,5)
```

```python
# Agar humpe sirf ik hi expression ho execute krne ke liye then we preffer to use lambda function 

sum = lambda a, b: a + b
```

```python
sum(12, 14)
```

```python
sub = lambda a, b: a - b
```

```python
sub(12, 10)
```

```python
mul = lambda a, b: a * b
```

```python
mul(2, 4)
```

```python
def even(n):
    if n % 2 == 0:
        return True
```

```python
even(2)
```

```python
even1 = lambda a: a % 2 == 0
even1(24)
```

```python
def addition(x, y, z):
    return x + y + z
```

```python
addition(2, 4, 6)
```

```python
add3 = lambda x, y, z: x + y + z
add3(2, 4, 6)
```
