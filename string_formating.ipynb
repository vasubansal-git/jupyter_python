```python
def greeting(name):
    return "Hello {}. Welcome to the community".format(name)
```

```python
greeting("Vasu")
```

```python
def welcome_email(firstname, lastname):
    return "Welcome {}.Your surname is {}".format(firstname, lastname) 
```

```python
welcome_email("Vasu", "Bansal")
```

```python
def details(name, age):
    return "Hello {}. Your age is {}".format(name, age)
```

```python
details("Vasu", 19)
```

```python
def generate_salary_slip(name, basic_salary):
    
    hra = basic_salary * 0.20
    da = basic_salary * 0.10
    total_salary = basic_salary + hra + da

    name = name.title()

    # Print formatted salary slip
    print("-" * 35)
    print("SALARY SLIP".center(35))
    print("-" * 35)
    print(f"{'Employee Name':<15} : {name}")
    print(f"{'Basic Salary':<15} : ₹{basic_salary:.2f}")
    print(f"{'HRA (20%)':<15} : ₹{hra:.2f}")
    print(f"{'DA (10%)':<15} : ₹{da:.2f}")
    print("-" * 35)
    print(f"{'Total Salary':<15} : ₹{total_salary:.2f}")
    print("-" * 35)


# Function Call
generate_salary_slip("vasu bansal", 25000)
```
