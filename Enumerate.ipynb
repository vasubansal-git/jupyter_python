# Enumerate function
## enumerate() is a built-in function that adds a counter (index) to an iterable (like list, tuple, string) and returns it as an enumerate object.It is mostly used in for loops when you need both index and value.

```python
# Enumerate function in python

a = [12, 56, 80, 90, 58]

index = 0
for x in a:
    print(x)
    if (index == 3):
        print("Vasu, Awesome")
    index += 1
```

```python
marks = [8, 34, 50, 90, 57, 10, 98, 99, 85]

for index, mark in enumerate(marks):
    print(mark)
    if(index == 7):
        print("Vasu, Awesome")
```

```python
fruits = ["apple", "banana", "mango"]

for index, value in enumerate(fruits, start = 1):
    print(index, value)
```

```python
string = "Vasu Bansal"

for index, char in enumerate(string, start = 1):
    print(index, char)
```
